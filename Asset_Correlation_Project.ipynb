```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# The 'Universe' - US Stocks, Gold, Bonds, Crypto, and Aussie Stocks
tickers = ["SPY", "GLD", "TLT", "BTC-USD", "VAS.AX"]
data = yf.download(tickers, start="2023-01-01", end="2025-12-25", auto_adjust=True)

# Fix the multi-header issue from your first screenshot
if isinstance(data.columns, pd.MultiIndex):
    data = data['Close']
```

```python
# Calculate % changes (Returns)
returns = data.pct_change().dropna()

# Calculate Correlation
corr = returns.corr()

# Create the Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='RdYlGn', center=0)
plt.title('Multi-Asset Correlation Matrix')
plt.show()
```
